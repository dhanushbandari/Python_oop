```python
class A:
    def show(self):
        print("Class A")

class B(A):
    def show_b(self):
        print("Class B")

class C(A):
    def show_c(self):
        print("Class C")

class D(B, C):
    def show_d(self):
        print("Class D")

# Example
d = D()
d.show()      # From A (resolved via MRO)
d.show_b()
d.show_c()
d.show_d()

# Display Method Resolution Order
print(D.mro())
```
