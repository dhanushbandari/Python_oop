```python
class Calculator:
    def add(self,a,b):
        return a+b
    def add (self,a,b=0):
        return a+b
    
c=Calculator()
print("Add one number:",c.add(5))
print("Add two number",c.add(5,10))
```

```python
class DataProcessor:
    def Process(self,*args):
        return sum(args)
D=DataProcessor()
print("Sum of numbers:",D.Process(5,10,20))
```

```python
class Person:
    def __init__(self,name,age=None):
        self.name=name
        if age is not None:
            self.age=age
        else:
            self.age="Age not specified"
    def display(self):
        print("Name:",self.name)
        print("Age:",self.age)

p1=Person("Alice",30)
p2=Person("Bob")
p1.display()
p2.display()
```

```python
from multipledispatch import dispatch
import math

class Area:
    
    @dispatch(int)
    def find_area(self, radius):
        return math.pi * radius * radius   # Circle
    
    @dispatch(int, int)
    def find_area(self, length, breadth):
        return length * breadth            # Rectangle


# Testing
a = Area()
print("Circle area:", a.find_area(5))
print("Rectangle area:", a.find_area(4, 6))
```

```python
class Configuration:
    def __init__(self, **kwargs):
        self.settings = kwargs   # store all values directly

    def display(self):
        print(self.settings)


# Testing
c1 = Configuration()
c2 = Configuration(host="localhost", port=8080, debug=True)

c1.display()   # {}
c2.display()   # {'host': 'localhost', 'port': 8080, 'debug': True}
```
